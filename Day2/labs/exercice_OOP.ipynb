## **Scenario: Library Management System**
You are tasked with building a **Library Management System** using Object-Oriented Programming in Python. The system should manage **Books** and **Users** with basic operations.

---

## **Exercise Instructions**

### **1️⃣ Create a `Book` Class**  
✅ Attributes:  
- `title` (string)  
- `author` (string)  
- `isbn` (string)  
- `available` (boolean, defaults to `True`)  

✅ Methods:  
- `__str__()`: Returns book details in a readable format.  
- `borrow()`: Marks the book as unavailable (`False`).  
- `return_book()`: Marks the book as available (`True`).  

**Example Usage**
```python
book1 = Book("Python Programming", "John Doe", "978-1234567890")
print(book1)  # Output: "Python Programming by John Doe (ISBN: 978-1234567890)"
book1.borrow()
print(book1.available)  # Output: False
book1.return_book()
print(book1.available)  # Output: True
```

---

### **2️⃣ Create a `User` Class**  
✅ Attributes:  
- `name` (string)  
- `user_id` (integer)  
- `borrowed_books` (list of books, empty by default)  

✅ Methods:  
- `borrow_book(book)`: Adds a book to `borrowed_books` if available.  
- `return_book(book)`: Removes a book from `borrowed_books`.  

**Example Usage**
```python
user1 = User("Alice", 101)
user1.borrow_book(book1)  # Alice borrows "Python Programming"
print(user1.borrowed_books)  # Output: ["Python Programming"]
user1.return_book(book1)
print(user1.borrowed_books)  # Output: []
```

---

### **3️⃣ Create a `Library` Class**
✅ Attributes:  
- `books` (list of all books in the library)  
- `users` (list of all users in the system)  

✅ Methods:  
- `add_book(book)`: Adds a new book to the library.  
- `add_user(user)`: Registers a new user.  
- `list_available_books()`: Prints all available books.  
- `borrow_book(user, book_title)`: Allows a user to borrow a book if available.  
- `return_book(user, book_title)`: Allows a user to return a book.  

**Example Usage**
```python
library = Library()
library.add_book(book1)
library.add_user(user1)
library.list_available_books()  # Should show "Python Programming"
library.borrow_book(user1, "Python Programming")
library.list_available_books()  # Should now be empty
library.return_book(user1, "Python Programming")
library.list_available_books()  # Should show the book again
```

### **4️⃣ Implement Inheritance**
✅ Create a subclass **`PremiumUser`** that inherits from `User`:  
- Allows borrowing **more than 3 books at a time**.  

**Example**
```python
premium_user = PremiumUser("Bob", 102)
premium_user.borrow_book(book1)  # Can borrow more than 3 books
```

### **5️⃣ Implement Polymorphism**
✅ Modify the `__str__()` method in `Book`, `User`, and `Library` to return useful information when printed.

## **Bonus Challenge **
🔹 Implement **a search function** to find books by title.  
🔹 Store book data in a JSON file and load it when the system starts.  

---

````python
import json
import os


# 1️⃣ Book Class
class Book:
    def __init__(self, title: str, author: str, isbn: str, available: bool = True):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.available = available

    def __str__(self):
        status = "Available" if self.available else "Borrowed"
        return f"{self.title} by {self.author} (ISBN: {self.isbn}) [{status}]"

    def borrow(self):
        if self.available:
            self.available = False
            return True
        return False

    def return_book(self):
        self.available = True


# 2️⃣ User Class
class User:
    MAX_BOOKS = 3

    def __init__(self, name: str, user_id: int):
        self.name = name
        self.user_id = user_id
        self.borrowed_books: list[Book] = []

    def __str__(self):
        books = ", ".join(b.title for b in self.borrowed_books) or "None"
        return f"User: {self.name} (ID: {self.user_id}) | Borrowed: [{books}]"

    def borrow_book(self, book: Book):
        if len(self.borrowed_books) >= self.MAX_BOOKS:
            print(f"{self.name} cannot borrow more than {self.MAX_BOOKS} books.")
            return False
        if not book.available:
            print(f"'{book.title}' is not available.")
            return False
        book.borrow()
        self.borrowed_books.append(book)
        return True

    def return_book(self, book: Book):
        if book in self.borrowed_books:
            book.return_book()
            self.borrowed_books.remove(book)
            return True
        print(f"{self.name} does not have '{book.title}'.")
        return False


# 3️⃣ Library Class
class Library:
    def __init__(self):
        self.books: list[Book] = []
        self.users: list[User] = []

    def __str__(self):
        return (
            f"Library: {len(self.books)} book(s), {len(self.users)} user(s)\n"
            f"Available: {len([b for b in self.books if b.available])}"
        )

    def add_book(self, book: Book):
        self.books.append(book)

    def add_user(self, user: User):
        self.users.append(user)

    def list_available_books(self):
        available = [b for b in self.books if b.available]
        if not available:
            print("No books available.")
        else:
            for book in available:
                print(f"  - {book}")

    def borrow_book(self, user: User, book_title: str):
        for book in self.books:
            if book.title == book_title:
                if user.borrow_book(book):
                    print(f"'{book_title}' borrowed by {user.name}.")
                return
        print(f"Book '{book_title}' not found in the library.")

    def return_book(self, user: User, book_title: str):
        for book in user.borrowed_books:
            if book.title == book_title:
                user.return_book(book)
                print(f"'{book_title}' returned by {user.name}.")
                return
        print(f"{user.name} does not have '{book_title}'.")

    # Bonus: Search function
    def search(self, query: str):
        results = [b for b in self.books if query.lower() in b.title.lower()]
        if not results:
            print(f"No books found matching '{query}'.")
        else:
            for book in results:
                print(f"  - {book}")
        return results

    # Bonus: Save to JSON
    def save_to_json(self, filepath: str = "library_data.json"):
        data = {
            "books": [
                {"title": b.title, "author": b.author, "isbn": b.isbn, "available": b.available}
                for b in self.books
            ],
            "users": [
                {"name": u.name, "user_id": u.user_id, "borrowed_books": [b.title for b in u.borrowed_books]}
                for u in self.users
            ],
        }
        with open(filepath, "w") as f:
            json.dump(data, f, indent=2)
        print(f"Library data saved to {filepath}.")

    # Bonus: Load from JSON
    def load_from_json(self, filepath: str = "library_data.json"):
        if not os.path.exists(filepath):
            print(f"File '{filepath}' not found.")
            return
        with open(filepath, "r") as f:
            data = json.load(f)
        self.books = [Book(**b) for b in data["books"]]
        self.users = []
        for u in data["users"]:
            user = User(u["name"], u["user_id"])
            for title in u["borrowed_books"]:
                for book in self.books:
                    if book.title == title:
                        user.borrowed_books.append(book)
            self.users.append(user)
        print(f"Library data loaded from {filepath}.")


# 4️⃣ PremiumUser (Inheritance)
class PremiumUser(User):
    MAX_BOOKS = 10  # Can borrow more than 3 books

    def __str__(self):
        books = ", ".join(b.title for b in self.borrowed_books) or "None"
        return f"Premium User: {self.name} (ID: {self.user_id}) | Borrowed: [{books}]"
````

````python
# === Demo / Testing ===

# Create books
book1 = Book("Python Programming", "John Doe", "978-1234567890")
book2 = Book("Clean Code", "Robert C. Martin", "978-0132350884")
book3 = Book("Design Patterns", "Gang of Four", "978-0201633610")
book4 = Book("The Pragmatic Programmer", "David Thomas", "978-0135957059")

print("=== Book Class ===")
print(book1)
book1.borrow()
print(f"After borrow -> available: {book1.available}")
book1.return_book()
print(f"After return  -> available: {book1.available}")

print("\n=== User Class ===")
user1 = User("Alice", 101)
user1.borrow_book(book1)
print(user1)
user1.return_book(book1)
print(user1)

print("\n=== Library Class ===")
library = Library()
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)
library.add_book(book4)
library.add_user(user1)

print("Available books:")
library.list_available_books()

library.borrow_book(user1, "Python Programming")
print("\nAfter borrowing 'Python Programming':")
library.list_available_books()

library.return_book(user1, "Python Programming")
print("\nAfter returning:")
library.list_available_books()

print("\n=== PremiumUser (Inheritance) ===")
premium_user = PremiumUser("Bob", 102)
library.add_user(premium_user)
for b in [book1, book2, book3, book4]:
    library.borrow_book(premium_user, b.title)
print(premium_user)

print("\n=== Search (Bonus) ===")
library.search("Python")

print("\n=== Save/Load JSON (Bonus) ===")
library.save_to_json("library_data.json")

# Load into a new library instance
library2 = Library()
library2.load_from_json("library_data.json")
print(library2)
````
